```python
import pandas as pd
import spotipy
import spotipy.util
import os
import pickle as pkl
import re
from importlib import reload

pd.options.display.max_columns = 999

% matplotlib inline
qs = __import__('query-spotify')
```

```python
# define data directory where credentials and data (old and new) are stored
data_dir = os.path.join('..','data')
with open(os.path.join(data_dir, 'credentials.dict'), 'rb') as f:
    credentials = pkl.load(f)

token = qs.get_auth_token(**credentials)
```

```python
def clean_artist(x):
    # help out Olivia Newton-John
    x = re.sub('newtonjohn', 'newton-john', x)
    
    # return only text that appears before 'featuring', otherwise
    # Spotify will not return a search result
    if re.findall('featuring', x):
        return x.split('featuring')[0]
    else:
        return x
    

df_orig = pd.read_csv(os.path.join(data_dir, 'billboard-lyrics.csv'), encoding='latin1')

df = df_orig.copy()
df.columns = df.columns.str.lower()
df['artist'] = df['artist'].apply(clean_artist)
df.head()
```

```python
# get relevant spotify features for each song
# a dataframe of songs that did not return a result are stored in 'failed'
df2, failed = qs.query_spotify(df, token, verbose=True)
```

```python
# merge data from billboard and spotify
df1 = df_orig.copy()
df1.columns = df1.columns.str.lower()
df1 = df1.reset_index()

df_merge = pd.merge(df1, df2, how='left', left_on='index',
                    right_on='billboard_df_index')
column_mask = df_merge.columns.str.contains('billboard')
df_merge = df_merge.loc[:, ~column_mask]
df_merge = df_merge.drop(['index', 'source', 'id', 'type', 'uri'], axis=1)
```

```python
# save resulting dataframes to data directory
df_merge.to_csv(os.path.join(data_dir, 'billboard-spotify.csv'), index=False)
failed.to_csv(os.path.join(data_dir, 'failed.csv'), index=False)
```
